```python
# HR Attrition Analysis 
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# For better visuals
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 4)


#  Load the Dataset

df = pd.read_csv("D:/powerbi/projects/HR-Employee-Attrition.csv")
```

```python
#  Basic Overview

print("Shape of dataset:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nMissing values:\n", df.isnull().sum())

print("\nData types:\n", df.dtypes)
print("\nSample rows:\n", df.head())
```

```python
#  Target Variable – Attrition
sns.countplot(x='Attrition', data=df, palette='Set2')
plt.title("Attrition Distribution")
plt.show()
```

```python
# Attrition % value

attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
print("\nAttrition Rate:\n", attrition_rate)
```

```python
#  Numerical Analysis

num_cols = ['Age', 'MonthlyIncome', 'DistanceFromHome', 'TotalWorkingYears', 'YearsAtCompany']

for col in num_cols:
    plt.figure()
    sns.boxplot(x='Attrition', y=col, data=df, palette='coolwarm')
    plt.title(f'{col} vs Attrition')
    plt.tight_layout()
    plt.show()
```

```python
#  Correlation Heatmap

df_corr = df.copy()
df_corr['Attrition'] = df_corr['Attrition'].map({'Yes': 1, 'No': 0})  # encode for correlation

plt.figure(figsize=(12, 8))
sns.heatmap(df_corr.corr(numeric_only=True), annot=True, cmap='YlGnBu', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#  Categorical vs Attrition

cat_cols = ['Gender', 'MaritalStatus', 'JobRole', 'Department', 'EducationField', 'OverTime']

for col in cat_cols:
    plt.figure(figsize=(6, 3))
    sns.countplot(data=df, x=col, hue='Attrition', palette='Set3')
    plt.title(f'{col} vs Attrition')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
#  Step 8: Key Insights

print("\n Insights:")
print("- Employees who work overtime are more likely to leave.")
print("- Younger employees and those with lower income show higher attrition.")
print("- Job roles like 'Sales Representative' and 'Laboratory Technician' have higher attrition.")
print("- High distance from home and low years at company also correlate with attrition.")
print("- Marital status shows that single employees tend to leave more.")
```
